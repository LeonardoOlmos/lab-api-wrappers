```python
# Create spotipy instance

sp = spotipy.Spotify(auth_manager= SpotifyClientCredentials(client_id = client_id, client_secret = client_secret))
```

```python
def get_song_info(songs):
    ids = []
    
    for i in range(50):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(50,100):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(100, 150):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(150, 200):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(200, 250):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(250, 300):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(300, 350):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(350, 400):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(400, 450):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(450, 500):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(500, 550):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(550, 600):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    time.sleep(35)
    for i in range(600, len(songs_df)):
        try :
            song_info = sp.search(q = songs[i], type = "track", limit = 1)
            ids.append(song_info['tracks']['items'][0]['id'])
        except:
            ids.append(np.NaN)
    
    return ids
```

```python
def get_features_by_song_id(songs_lst):
    features = []
            
    for i in range(50):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(50,100):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(100, 150):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(150, 200):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(200, 250):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(250, 300):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(300, 350):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(350, 400):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(400, 450):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(450, 500):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(500, 550):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(550, 600):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
    time.sleep(35)
    for i in range(600, len(songs_df)):
        try :
            features.append(sp.audio_features(songs_lst[i]))
        except:
            features.append([])
        
    return features
```

```python
def get_songs_dataframe(songs_df, features_df):
    merged_df = pd.concat([songs_df, features_df], axis = 1)
    
    return merged_df
```
